# Plot eigenvalues decay of Poisson's ratio inversion as a function of data noise, number of surface observations, and earthquake depth

This notebook plots the eigenvalues decays of several Poisson's ratio inversion by changing the noise level in the data, the number of surface geodetic stations, and the depth of the earthquake.


```python
# Load libraries
import numpy as np
import pandas as pd
# Load libraries for plotting
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
from matplotlib.collections import LineCollection
from matplotlib import rc
rc('text', usetex=True)
rc('font', family='serif', size=16)
rc('legend', fontsize=16)
TINY_SIZE = 14
SMALL_SIZE = 20
MEDIUM_SIZE = 30
BIGGER_SIZE = 40
plt.rc('font', size=SMALL_SIZE)
plt.rc('axes', titlesize=SMALL_SIZE)
plt.rc('axes', labelsize=SMALL_SIZE)
plt.rc('xtick', labelsize=SMALL_SIZE)
plt.rc('ytick', labelsize=SMALL_SIZE)
plt.rc('legend', fontsize=SMALL_SIZE)
plt.rc('figure', titlesize=BIGGER_SIZE)
```

```python
# Load eigenvalues from noise and Nobs
# Load file of different data noise
model_name = "eigenvalues_noise_mu_nu.txt"
noise_m = np.array([ 1.00e-04, 2.00e-03, 2.20e-03, 2.75e-03, 3.50e-03, 5.00e-03,
                   6.50e-03, 8.00e-03, 1.00e-02, 1.20e-02, 1.50e-02, 3.00e-02,
                   5.00e-02, 7.50e-02, 1.00e-01, 1.25e-01, 1.50e-01, 2.00e-01,
                   3.00e-01 ])
MAX_u = 3.0559
noise_perc = noise_m/MAX_u*100
lmbda_noise = pd.read_csv(datadir + model_name, delimiter=' ')

# Load file of different numbers of surface observations
model_name = "eigenvalues_Nobs_mu_nu.txt"
Nobs = np.array([ 1, 5, 11, 17, 23,  
                 27, 33, 37, 45, 50,  
                 56, 61, 67, 72, 78,  
                 84, 94, 101, 112, 124, 
                 145, 172, 220])
lmbda_Nobs = pd.read_csv(datadir + model_name, delimiter=' ')

# Load file of different earthquake depths
model_name = "eigenvalues_EQdepth_mu_nu.txt"
EQdepth = np.array([ 5, 10, 15, 20, 25, 
                    30, 35, 40, 45, 50, 
                    55, 60 ])
lmbda_EQdepth = pd.read_csv(datadir + model_name, delimiter=' ')
```

```python
# Compute the metric LDPDH, which is the log-determinant of the pre-conditioned data misfit Hessian
lmbda_0 = 1e0
LDPDH_noise = np.array( 0.5*np.sum( np.log(1.+lmbda_noise[(lmbda_noise > lmbda_0)] ) ) )
LDPDH_Nobs = np.array( 0.5*np.sum( np.log(1.+lmbda_Nobs[(lmbda_Nobs > lmbda_0)] ) ) )
LDPDH_EQdepth = np.array( 0.5*np.sum( np.log(1.+lmbda_EQdepth[(lmbda_EQdepth > lmbda_0)] ) ) )
```

```python
# For labels of subplots (a,b,c,d)
props = dict(boxstyle='round', lw=0.25, facecolor='ivory', alpha=0.95)

colors_noise = plt.cm.Blues_r( np.linspace(0., 1., lmbda_noise.shape[1]) )
colors_Nobs = plt.cm.Reds( np.linspace(0., 1., lmbda_Nobs.shape[1]) )
colors_EQdepth = plt.cm.Greens_r( np.linspace(0., 1., lmbda_EQdepth.shape[1]) )
# Decide parameters for plotting
lww = 2.
lww_inset = 4.
```

```python
# Plot
fig, axi = plt.subplots(1, 3, figsize=(16,5.5), sharex=True, sharey=True)
plt.subplots_adjust(wspace=0.1)

# Plot left
ax1 = axi[0]
n_lmbda = lmbda_noise.shape[0]
ax1.plot(range(1, n_lmbda+1), np.ones(n_lmbda), linestyle="dotted", color="gray", zorder=1)
for i in range(lmbda_noise.shape[1]):
    ax1.plot(range(1, n_lmbda+1), lmbda_noise.iloc()[:,i], ls='solid', lw=lww, color=colors_noise[i],
             zorder=1)
ax1.set_xlim((1, 30))
ax1.set_ylim((1e-4, 1e10))
ax1.set_yscale('log')
ax1.set_xlabel(r"$i$")
ax1.set_ylabel(r"$\lambda_i\, (\tilde{\mathbf{H}}^{\mathrm{misfit}})$")
ax1.text(0.04, 0.09, r'\textbf{(a)}', transform=ax1.transAxes, fontsize=16,
                verticalalignment='top', bbox=props) 

# Plot inset
axins1 = ax1.inset_axes([0.5, 0.7, 0.5, 0.3])
axins1.plot(noise_perc, LDPDH_noise,'k')
points_noise = np.array([noise_perc, LDPDH_noise]).T.reshape(-1, 1, 2)
segments_noise = np.concatenate([points_noise[:-1], points_noise[1:]], axis=1)
lc_noise = LineCollection(segments_noise, cmap='Blues_r')
lc_noise.set_array(np.linspace(0., 1., lmbda_noise.shape[1]))
lc_noise.set_linewidth(lww_inset)
axins1.add_collection(lc_noise)
axins1.set_xlabel(r'data noise [\%]', fontsize=14)
axins1.set_ylabel(r'LDPDH', fontsize=14)

axins1.set_ylim([-5, 165])
axins1.set_xlim([-0.5, 10.75])
axins1.tick_params(axis='both', direction='out', labelsize=14)
axins1.set_xticks([0, 5, 10])
axins1.set_yticks([0, 50, 100, 150])


# Plot middle 
ax2 = axi[1]
n_lmbda = lmbda_Nobs.shape[0]
ax2.plot(range(1, n_lmbda+1), np.ones(n_lmbda), linestyle="dotted", color="gray", zorder=1)
for i in range(lmbda_Nobs.shape[1]):
    ax2.plot(range(1, n_lmbda+1), lmbda_Nobs.iloc()[:,i], ls='solid', lw=lww, color=colors_Nobs[i],
             zorder=1)
ax2.set_yscale('log')
ax2.set_xlabel(r"$i$")
ax2.text(0.04, 0.09, r'\textbf{(b)}', transform=ax2.transAxes, fontsize=16,
                verticalalignment='top', bbox=props) 

# Plot inset
axins2 = ax2.inset_axes([0.5, 0.7, 0.5, 0.3])
axins2.plot(Nobs, LDPDH_Nobs,'k')
points_Nobs = np.array([Nobs, LDPDH_Nobs]).T.reshape(-1, 1, 2)
segments_Nobs = np.concatenate([points_Nobs[:-1], points_Nobs[1:]], axis=1)
lc_Nobs = LineCollection(segments_Nobs, cmap='Reds')
lc_Nobs.set_array(np.linspace(0., 1., lmbda_Nobs.shape[1]))
lc_Nobs.set_linewidth(lww_inset)
axins2.add_collection(lc_Nobs)
axins2.set_xlabel(r'$N^\circ$ obs', fontsize=14)
axins2.set_ylabel(r'LDPDH', fontsize=14)

axins2.set_ylim([0, 130])
axins2.tick_params(axis='both', direction='out', labelsize=14)
axins2.set_xticks([0, 100, 200])


# Plot right panel
ax3 = axi[2]
n_lmbda = lmbda_Nobs.shape[0]
ax3.plot(range(1, n_lmbda+1), np.ones(n_lmbda), linestyle="dotted", color="gray", zorder=1)
for i in range(lmbda_EQdepth.shape[1]):
    ax3.plot(range(1, n_lmbda+1), lmbda_EQdepth.iloc()[:,i], ls='solid', lw=lww, color=colors_EQdepth[i],
             zorder=1)
ax3.set_yscale('log')
ax3.set_xlabel(r"$i$")
ax3.text(0.04, 0.09, r'\textbf{(c)}', transform=ax3.transAxes, fontsize=16,
                verticalalignment='top', bbox=props) 

# Plot inset
axins3 = ax3.inset_axes([0.5, 0.7, 0.5, 0.3])
axins3.plot(EQdepth, LDPDH_EQdepth, 'k')
points_EQdepth = np.array([EQdepth, LDPDH_EQdepth]).T.reshape(-1, 1, 2)
segments_EQdepth = np.concatenate([points_EQdepth[:-1], points_EQdepth[1:]], axis=1)
lc_EQdepth = LineCollection(segments_EQdepth, cmap='Greens_r')
lc_EQdepth.set_array(np.linspace(0., 1., lmbda_EQdepth.shape[1]))
lc_EQdepth.set_linewidth(lww_inset)
axins3.add_collection(lc_EQdepth)
axins3.set_xlabel(r'EQ depth [km]', fontsize=14)
axins3.set_ylabel(r'LDPDH', fontsize=14)

axins3.set_ylim([35, 90])
axins3.tick_params(axis='both', direction='out', labelsize=14)
axins3.set_xticks([0, 25, 50])


plt.show()
```
